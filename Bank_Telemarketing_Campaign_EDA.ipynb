## Bank Telemarketing Campaign Case Study.

This case study explores the importance of Exploratory Data Analysis (EDA) as a key process of  machine learning through the example of a bank marketing campaign, demonstrating why EDA is a crucial initial step in the machine learning process.

#### Problem Statement:



The bank offers financial products like savings accounts, current accounts, and debit cards. To boost revenue, it conducts marketing campaigns for products such as credit cards, term deposits, and loans, targeting existing customers. These campaigns must be cost-efficient to maximize both revenue and profit.EDA is applied to the dataset to analyze patterns and provide insights for future marketing campaigns.

The bank conducted a telemarketing campaign to promote term deposits, aiming to strengthen long-term relationships with existing customers. The dataset includes information on all customers contacted during the year to open term deposit accounts.


**What is the term Deposit?**

Term deposits, or fixed deposits, are cash investments for periods ranging from 1 month to 5 years at fixed interest rates, which are higher than savings account rates. The total amount, including interest, is received at maturity. Early withdrawal incurs a penalty and forfeits interest returns.

**Aim**

The aim of this study is to conduct an end-to-end exploratory data analysis (EDA) on a bank telemarketing campaign dataset. The primary goal is to identify areas where the bank can enhance its positive response rate by understanding the factors influencing customers' decision-making processes.

#### Dividing "jobedu" column into job and education categories.

```python
#Extract job in newly created 'job' column from "jobedu" column.

# Define a function to extract job from 'jobedu' string
def extract_job(jobedu):
# Split the string by comma (',') and return the first part (job)
    return jobedu.split(',')[0].strip()

# Apply the function to create a new 'job' column containing only job titles
inp0['job'] = inp0['jobedu'].apply(extract_job)

# Display the updated DataFrame with 'job' column extracted
inp0.head()
```

```python
#Extract education in newly created 'education' column from "jobedu" column.

# Define a function to extract education from 'jobedu' string
def extract_education(jobedu):
    # Split the string by comma (',') and return the second part (education)
    return jobedu.split(',')[1].strip() if ',' in jobedu else ''

# Apply the function to create a new 'education' column
inp0['education'] = inp0['jobedu'].apply(extract_education)

inp0.head()
```
